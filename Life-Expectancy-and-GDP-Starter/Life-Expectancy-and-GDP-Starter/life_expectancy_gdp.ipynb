```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
life = pd.read_csv("all_data.csv")
life.head(10)
```

```python
life.info()
```

```python
print(life.Country.unique(), life.Year.unique())
```

```python
sns.histplot(life.GDP)
plt.show()
plt.clf()
```

```python
sns.histplot(life['Life expectancy at birth (years)'])
plt.show()
plt.clf()
```

```python
sns.boxplot(data = life, x = 'Country', y = 'Life expectancy at birth (years)')
plt.xticks(rotation = 40)
plt.title('Life expectancy at birth (years)')
plt.show()
plt.clf()
```

```python
sns.boxplot(data = life, x = 'Country', y = 'GDP')
plt.xticks(rotation = 40)
plt.title('GDP')
plt.show()
plt.clf()
```

```python
sns.lineplot(data = life, x = 'Year', y = 'GDP', hue = 'Country')
plt.show()
plt.clf()
```

```python
graphGDP = sns.FacetGrid(life, col = 'Country', hue ='Country', col_wrap = 3, sharey=False,)
graphGDP.map(sns.lineplot, "Year",'GDP').add_legend()
graphGDP
```

```python
graphExp = sns.FacetGrid(life, col = 'Country', hue ='Country', col_wrap = 3, sharey=False,)
graphExp.map(sns.lineplot, "Year",'Life expectancy at birth (years)').add_legend()
graphExp
```

```python
sns.lmplot(data = life , x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Country',fit_reg=False)
plt.show()
plt.clf()
```
